```python
import networkx as nx

import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
M=nx.read_gml("/Users/adriancavallaris/Documents/polblogs.gml")
```

```python
# create weighted graph from M
G = nx.Graph()
for u,v,data in M.edges_iter(data=True):
    if G.has_edge(u,v):
        G[u][v]['weight'] += 1
    else:
        G.add_edge(u, v, weight=1)
```

### 1. Create a nice looking graph of the political blog network

I tried to play around the sizes on here, and the smaller ones (40, 40) blended together and looked awful, so I went with (200, 200), since I can zoom in and see the connections and labels. 

```python
plt.figure(figsize=(200,200))
pos = nx.spring_layout(G, iterations = 1000, scale = 5, k = .3)
nx.draw_networkx_labels(G, pos, fontsize = 8)
font = {'fontname'   : 'Helvetica',
            'color'      : 'k',
            'fontweight' : 'bold',
            'fontsize'   : 14}
plt.title("Political Blog Network", font)


nx.draw_networkx_nodes(G,pos, node_color ='b', alpha = 0.4)
nx.draw_networkx_edges(G,pos, alpha = 0.4, node_size = 0, width = 1, edge_color = 'k')
plt.axis('off')
plt.show()
```

```python
centrality = nx.eigenvector_centrality_numpy(G)
```

### 2. What are the 10 most influential blogs?

```python
blogList = []
for node in centrality:
    blogList.append((node,centrality[node]))
    
sBlogList = blogList.sort(key=lambda x: x[1])
```

```python
sBlogList = sorted(blogList, key=lambda x: x[1], reverse=True)
```
